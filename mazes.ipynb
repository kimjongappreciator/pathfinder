```
def DFSsolver(maze, destination, start, path=[], passed=[]):
    (x,y) = start 

    

    if (x,y) not in passed:
        path.append((x, y))
        passed.append((x, y))

    
    if (x,y) == destination:
        return path

    
    else:
        pointqueue = [(x - 1, y), (x, y - 1), (x + 1, y), (x, y + 1)]
        queue = []
        for point in pointqueue:
           
            if len(maze)-1 >= point[1] >= 0 and len(maze[0])-1 >= point[0] >= 0:
                
                if maze[point[1]][point[0]] == " " and point not in passed:
                    queue.append(point)

       
        if len(queue)!= 0:
            for point in queue:
                return DFSsolver(maze, destination, point, path)

       
        else:
            path.remove((x, y))
            return DFSsolver(maze, destination, path[-1], path)
```

```
def bruteforce(deck, player):
    cola = []
    maze = deck
    counter = 0
    j = player.y
    i = player.x
    

    while True:
        
        maze[i][j] = 1
        if maze[i][j+1] == "X":
            break
        elif maze[i][j-1] == "X":
            break
        elif maze[i+1][j] == "X":
            break
        elif maze[i-1][j] == "X":
            break
            
        elif maze[i][j+1] == " ":
            #print("d")
            cola.append("d")
            counter +=1
            j += 1
        elif maze[i][j-1] == " ":
            #print("u")
            cola.append("u")
            counter+=1
            j -=1
        elif maze[i+1][j] == " ":
            #print("r")
            cola.append("r")
            counter+=1
            i+=1
        elif maze[i-1][j] == " ":
            #print("l")
            cola.append("l")
            counter+=1
            i-=1
    
                 
    print(cola)
    return cola
```
